```python
from utils.agent import *
from utils.dataset import read_voc_dataset
import tqdm.notebook as tq
import pickle

_, val_loader2007 = read_voc_dataset(path="./data/PascalVOC2012", year='2012', download=True)
```

```python
datasets_per_class = sort_class_extract([val_loader2007])
```

```python
torch.cuda.empty_cache()
results = {}
for i in classes:
    results[i] = []
model_name='vgg16_3step'
for i in tq.tqdm(range(len(classes))):
    classe = classes[i]
    print("Class "+str(classe)+"...")
    agent = Agent_3alpha(classe, load=True, model_name=model_name)
    res = agent.evaluate(datasets_per_class[classe])
    results[classe] = res
```

```python
file_name = 'classes_results_' + model_name + '.pickle'
with open(file_name, 'wb') as handle:
    pickle.dump(results, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
